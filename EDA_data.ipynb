```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Typical donation amounts and Students reached 
sdf = df.sample(1000, random_state = 2025)
plt.scatter(sdf['students_reached'],sdf['donation_total'],color='red')
plt.xlabel('Students reached')
plt.ylabel('Donation total')
plt.title('Typical distribution of Total Donation Vs Students reached')
%matplotlib inline
```

```python
#Checking for trend in donation with time
date_donation = sdf[['date_posted','donation_total']]
date_donation['date_posted'] = pd.to_datetime(date_donation['date_posted'])

date_donation = date_donation.set_index('date_posted')

annual_donation = date_donation['donation_total'].resample('Y').sum()

annual_donation.plot(kind='line')
plt.xlabel('Date of posting project')
plt.ylabel('Total Donation')
plt.title('Trend of donation with time of project posting')
plt.tight_layout()
```

```python
# Convert 'date_posted' to datetime
sdf['date_posted'] = pd.to_datetime(sdf['date_posted'])

# Extract the year from the 'date_posted'
sdf['year'] = sdf['date_posted'].dt.year

# Group by year and 'fully_funded' to count projects funded vs not
funding_counts = sdf.groupby(['year', 'fully_funded']).size().unstack(fill_value=0)

# Plot the bar chart
funding_counts.plot(kind='bar', figsize=(8, 6))
plt.xlabel('Year')
plt.ylabel('Number of Projects')
plt.title('Number of Projects Funded vs Not Funded Each Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Checking for school_city and whether or not they were funded
city_fund = pd.crosstab(sdf['poverty_level'],sdf['fully_funded'])

plt.figure(figsize=(10,5))
city_fund.plot(kind = 'bar')
plt.xlabel('Poverty_level')
plt.ylabel('Count')
plt.title('Count of funding by poverty_level')
plt.tight_layout()
```

```python
#Checking for school_city and whether or not they were funded
city_fund = pd.crosstab(sdf['resource_type'],sdf['fully_funded'])

plt.figure(figsize=(10,5))
city_fund.plot(kind = 'bar')
plt.xlabel('Resource type')
plt.ylabel('Count')
plt.title('Count of funding by resource type')
plt.tight_layout()
```

```python
#Checking for school_city and whether or not they were funded
city_fund = pd.crosstab(sdf['school_state'],sdf['fully_funded'])

plt.figure(figsize=(14,5))
city_fund.plot(kind = 'bar')
plt.xlabel('States')
plt.ylabel('Count')
plt.title('Count of funding by states')
plt.tight_layout()
```
